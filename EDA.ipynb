```python
# make a new dataframe with raw_dataset.txt
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open('raw_dataset.txt') as f:
        data = f.readlines()
        clean_data = [x.strip() for x in data]
clean_list = [clean_data[i:i+8] for i in range(0, len(data), 8)]
```

```python
# make a new dataframe with raw_dataset.txt
df_new = pd.DataFrame(clean_list, columns=['date', 'competition', 'season', 'matchday', 'home_team', 'score', 'away_team', 'tactic'])
df_new
```

```python
# split score into home_score and away_score
df_new[['home_score', 'away_score']] = df_new.score.str.split(':', expand=True)
df_new
```

```python
arranged_df = df_new[['date', 'competition', 'season', 'matchday', 'home_team', 'home_score', 'away_score', 'away_team', 'tactic']]
arranged_df
```

```python
# remove the ' in all the columns
cleaned_df = arranged_df.replace("'", "", regex=True)
cleaned_df
```

```python
# changed the data type of home_score and away_score to int
cleaned_df['home_score'] = cleaned_df['home_score'].astype(int)
cleaned_df['away_score'] = cleaned_df['away_score'].astype(int)
cleaned_df.dtypes
```

```python
cleaned_df
```

```python
# check if Liverpool/Dortmund/Mainz won
# create a new column called 'win'
# if Liverpool won/Dortmund/Mainz, win = 1
# if Liverpool lost/Dortmund/Mainz, win = 0
# if Liverpool drew/Dortmund/Mainz, win = 0.5
# if the match is not Liverpool's, win = None
cleaned_df['win'] = None
for i in range(len(cleaned_df)):
    if cleaned_df['home_team'][i] == 'Liverpool' or cleaned_df['home_team'][i] == 'Bor. Dortmund' or cleaned_df['home_team'][i] == '1.FSV Mainz 05':
        if cleaned_df['home_score'][i] > cleaned_df['away_score'][i]:
            cleaned_df['win'][i] = 1
        elif cleaned_df['home_score'][i] < cleaned_df['away_score'][i]:
            cleaned_df['win'][i] = 0
        else:
            cleaned_df['win'][i] = 0.5
    elif cleaned_df['away_team'][i] == 'Liverpool' or cleaned_df['away_team'][i] == 'Bor. Dortmund' or cleaned_df['away_team'][i] == '1.FSV Mainz 05':
        if cleaned_df['away_score'][i] > cleaned_df['home_score'][i]:
            cleaned_df['win'][i] = 1
        elif cleaned_df['away_score'][i] < cleaned_df['home_score'][i]:
            cleaned_df['win'][i] = 0
        else:
            cleaned_df['win'][i] = 0.5
    else:
        cleaned_df['win'][i] = None
```

```python
cleaned_df.tail(20)
```

```python
# count the number of wins
no_wins = cleaned_df['win'].value_counts()
no_wins
```

```python
# count the number of wins for liverpool only
# save the number of wins in a variable called liv_wins
liv_wins = 0
liv_loss = 0
liv_draws = 0
liv_home = 0
liv_away = 0
liv_home_wins = 0
liv_home_losses = 0
liv_away_wins = 0
liv_away_losses = 0
for i in range(len(cleaned_df)):
    if cleaned_df['home_team'][i] == 'Liverpool':
        liv_home += 1
        if cleaned_df['home_score'][i] > cleaned_df['away_score'][i]:
            liv_wins += 1
            liv_home_wins += 1
        elif cleaned_df['home_score'][i] < cleaned_df['away_score'][i]:
            liv_loss += 1
            liv_home_losses += 1
        else:
            liv_draws += 1
    elif cleaned_df['away_team'][i] == 'Liverpool':
        liv_away += 1
        if cleaned_df['home_score'][i] > cleaned_df['away_score'][i]:
            liv_wins += 1
            liv_away_wins +=1
        elif cleaned_df['home_score'][i] < cleaned_df['away_score'][i]:
            liv_loss += 1
            liv_away_losses += 1
        else:
            liv_draws += 1
print("Liverpool played {} matches".format(liv_home + liv_away))
print("Liverpool won {} matches at home and lost {} matches at home".format(liv_home_wins, liv_home_losses))
print("Liverpool won {} matches away and lost {} matches away".format(liv_away_wins, liv_away_losses))
print("Liverpool won", liv_wins, "matches")
print("Liverpool lost", liv_loss, "matches")
print("Liverpool drew", liv_draws, "matches")
```

```python
# matches won at home vs matches won away
plt.pie([liv_home_wins, liv_away_wins], labels=['Home', 'Away'], autopct='%1.1f%%', startangle=90)
plt.show()
```

```python
# matches won away vs matches lost away
pie_labels = ['Won', 'Lost']
plt.pie([liv_away_wins, liv_away_losses], labels=['Won', 'Lost'], autopct='%1.1f%%', startangle=90)
plt.legend(pie_labels, loc="best")
plt.show()
```

```python
# change date to datetime
cleaned_df["date"] = pd.to_datetime(cleaned_df["date"])
cleaned_df.dtypes
```

```python

# Find how many matches Liverpool played in each competition
Epl_count = 0
Ucl_count = 0
Fa_count = 0
shield_count = 0
europa_count = 0
Efl_count = 0
leaguecup_count =0
for i in range(len(cleaned_df)):
    if cleaned_df['home_team'][i] == 'Liverpool' or cleaned_df['away_team'][i] == 'Liverpool':
        if cleaned_df['competition'][i] == 'UEFA Champions League':
            Ucl_count += 1
        elif cleaned_df['competition'][i] == 'Premier League':
            Epl_count += 1
        elif cleaned_df['competition'][i] == 'FA Cup':
            Fa_count += 1
        elif cleaned_df['competition'][i] == 'Community Shield':
            shield_count += 1
        elif cleaned_df['competition'][i] == 'Europa League':
            europa_count += 1
        elif cleaned_df['competition'][i] == 'EFL Cup':
            Efl_count += 1
        elif cleaned_df['competition'][i] == 'League Cup':
            leaguecup_count += 1



print("Liverpool played {} matches in the Premier League".format(Epl_count))
print("Liverpool played {} matches in the UEFA Champions League".format(Ucl_count))
print("Liverpool played {} matches in the FA Cup".format(Fa_count))
print("Liverpool played {} matches in the Community Shield".format(shield_count))
print("Liverpool played {} matches in the Europa League".format(europa_count))
print("Liverpool played {} matches in the EFL Cup".format(Efl_count))
print("Liverpool played {} matches in the League Cup".format(leaguecup_count))
```

```python
# Uefa Champions League
# Number of wins/losses/draws
liv_ucl_wins = 0
liv_ucl_loss = 0
liv_ucl_draws = 0
liv_ucl_home_wins = 0
liv_ucl_home_losses = 0
liv_ucl_draws = 0
liv_ucl_away_losses = 0
liv_ucl_away_wins = 0
for i in range(len(cleaned_df)):
    if cleaned_df['home_team'][i] == 'Liverpool' or cleaned_df['away_team'][i] == 'Liverpool':
        if cleaned_df['competition'][i] == 'UEFA Champions League':
            if cleaned_df['home_team'][i] == 'Liverpool':
                if cleaned_df['home_score'][i] > cleaned_df['away_score'][i]:
                    liv_ucl_wins += 1
                    liv_ucl_home_wins += 1
                elif cleaned_df['home_score'][i] < cleaned_df['away_score'][i]:
                    liv_ucl_loss += 1
                    liv_ucl_home_losses += 1
                else:
                    liv_ucl_draws += 1
            elif cleaned_df['away_team'][i] == 'Liverpool':
                if cleaned_df['home_score'][i] > cleaned_df['away_score'][i]:
                    liv_ucl_loss += 1
                    liv_ucl_away_losses += 1
                elif cleaned_df['home_score'][i] < cleaned_df['away_score'][i]:
                    liv_ucl_wins += 1
                    liv_ucl_away_wins += 1
                else:
                    liv_ucl_draws += 1


print("Liverpool won {} matches in the Uefa Champions League".format(liv_ucl_wins))
print("Liverpool lost {} matches in the Uefa Champions League".format(liv_ucl_loss))
print("Liverpool drew {} matches in the Uefa Champions League".format(liv_ucl_draws))
print("Liverpool won {} matches at home in the Uefa Champions League".format(liv_ucl_home_wins))
print("Liverpool lost {} matches at home in the Uefa Champions League".format(liv_ucl_home_losses))
print("Liverpool won {} matches away in the Uefa Champions League".format(liv_ucl_away_wins))
print("Liverpool lost {} matches away in the Uefa Champions League".format(liv_ucl_away_losses))
```

```python
# pie chart for Uefa Champions League
pie_labels_ucl = ['Won', 'Lost', 'Draw']
plt.pie([liv_ucl_wins, liv_ucl_loss, liv_ucl_draws], labels=pie_labels_ucl, autopct='%1.1f%%', startangle=90)
plt.legend(pie_labels_ucl, loc="best")
plt.show()
```

```python
# pie chart for Uefa Champions League home
pie_labels_ucl_home = ['Won', 'Lost']
plt.pie([liv_ucl_home_wins, liv_ucl_home_losses], labels=pie_labels_ucl_home, autopct='%1.1f%%', startangle=90)
plt.legend(pie_labels_ucl_home, loc="best")
plt.show()
```

```python
# pie chart for Uefa Champions League away
pie_chart_ucl_away = ['Won', 'Lost']
plt.pie([liv_ucl_away_wins, liv_ucl_away_losses], labels=pie_chart_ucl_away, autopct='%1.1f%%', startangle=90)
plt.legend(pie_chart_ucl_away, loc="best")
plt.show()
```
